```python
import sys, os, json, re
sys.path.append(r'C:\Users\pierr\Desktop\Polimi\delay_mdp\dmdp\notebooks')
from matplotlib import pyplot as plt
# import loaders as load
import numpy as np
import pandas as pd
import torch
```

```python

def stats(path, max_epoch=None):
    subfolders = [f.path for f in os.scandir(path) if f.is_dir()]
    mean = {}
    std = {}
    for d in subfolders:
        for i, ds in enumerate(os.listdir(os.path.join(path,d))):
            if i==0:
                with open(os.path.join(path,d,ds,'model_parameters.txt')) as f:
                    delay = json.loads(f.read())['delay']
                print(delay)
                mean[delay] = []
#                 std[delay] = []
            try:
                model_path = os.path.join(path,d,ds,'model_{}.pt'.format(max_epoch))
                model = torch.load(model_path)
            except:
                model_path = os.path.join(path,d,ds,'model.pt')
                model = torch.load(model_path)

            mean[delay].append(torch.as_tensor(model['avg_reward']).numpy())
#             std[delay].append(torch.as_tensor(model['std_reward']).numpy())
        std[delay] = np.stack(mean[delay]).std(0)
        mean[delay] = np.stack(mean[delay]).mean(0)
    
    return mean, std
```

```python
delay = 5
def lighten_color(color, amount=0.5):
    """
    Lightens the given color by multiplying (1-luminosity) by the given amount.
    Input can be matplotlib color string, hex string, or RGB tuple.

    Examples:
    >> lighten_color('g', 0.3)
    >> lighten_color('#F034A3', 0.6)
    >> lighten_color((.3,.55,.1), 0.5)
    """
    import matplotlib.colors as mc
    import colorsys
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = colorsys.rgb_to_hls(*mc.to_rgb(c))
    return colorsys.hls_to_rgb(c[0], 1 - amount * (1 - c[1]), c[2])


fig, ax = plt.subplots(1,1,figsize=(8,7))
fig.patch.set_facecolor('white')

def plot_ret(y_mean, y_std=None, ax=None, label=None, linestyle='solid', marker='o', color=None): 
    if color is None:
        p = ax.plot(y_mean, label=label, lw=1, ms=9, mec='black', linestyle=linestyle, marker=marker,markevery=1)
    else:
        p = ax.plot(y_mean, label=label, lw=1, ms=9, mec='black', linestyle=linestyle, marker=marker,markevery=1, c=color)
    if y_std is not None:
        ax.fill_between(x=np.arange(len(y_mean)), y1=y_mean+y_std, y2=y_mean-y_std, color=p[-1].get_color(), alpha=0.18)
    
    
linestyles = ['solid','dotted','dashed','dashdot',(0, (3, 5, 1, 5, 1, 5))]
markers = ['o','v','s','P','d']
    
# if colors:
#     colors = sns.color_palette("cubehelix", 6)
# else:
#     darken = [1.4,1.3,1.2,1.2,1.1]
#     colors = [lighten_color(plt.rcParams['axes.prop_cycle'].by_key()['color'][i],darken[i]) for i in range(5)]          
    
#
for k,v in models.items():
    plot_ret(models[k]['mean'][delay], models[k]['mean'][delay], ax=ax, marker=None, label=k)


ax.legend(fontsize='medium')
ax.yaxis.grid(True)
ax.xaxis.grid(True)

# delays = np.array([k for k in returns.keys()])
# ticks = ax.get_xticks().astype(int)
# a = np.zeros(len(np.where(ticks<0)[0]))
# b = np.zeros(len(np.where(ticks>len(delays))[0]))
# ticks = ticks[(ticks>=0) & (ticks<=len(delays))]
# xlabels = np.hstack((a, delays[ticks], b))
# ax.set_xticklabels(xlabels.astype(int))

# ax.set_title('Comparison on 5-Steps Delay')
ax.set_xlabel('Epochs',fontsize=15)
ax.set_ylabel('Return',fontsize=15)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)
plt.show()
```
